# Age Distribution

```python
age_man20 = (SA20[SA20['Q2']=='Man']['Q1'].value_counts().to_frame().reset_index()
           .rename(columns={'index':'Age', 'Q1':'Count'})) 
age_man20
```

```python
age_man21 = (SA21[SA21['Q2']=='Man']['Q1'].value_counts().to_frame().reset_index()
           .rename(columns={'index':'Age', 'Q1':'Count'})) 
age_man21
```

```python
age_woman20 = (SA20[SA20['Q2']=='Woman']['Q1'].value_counts().to_frame().reset_index()
           .rename(columns={'index':'Age', 'Q1':'Count'})) 
age_woman20
```

```python
age_woman21 = (SA21[SA21['Q2']=='Woman']['Q1'].value_counts().to_frame().reset_index()
           .rename(columns={'index':'Age', 'Q1':'Count'})) 
age_woman21
```

```python
answer1men = (pd.merge(age_man20,age_man21,how='outer', on='Age')
              .sort_values('Age', ascending = False)
              .rename(columns={'Count_x':'2020', 'Count_y':'2021'}))
answer1men.fillna(0, inplace=True)
answer1men
```

```python
answer1women = (pd.merge(age_woman20,age_woman21,how='outer', on='Age')
                .sort_values('Age', ascending = False)
                .rename(columns={'Count_x':'2020', 'Count_y':'2021'})) 
answer1women
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10), constrained_layout=True);
# fig.tight_layout();
fig.suptitle('Age Distribution' , color = 'lightpink', weight = 'bold');

# Men Age in 2020
sns.barplot(ax=axes[0,0], data=age_man20, x='Age', y='Count', color='#ffab40');
axes[0,0].set_title('Men Age in 2020',color = 'lightpink', weight = 'bold');
axes[0,0].grid(axis='y', lw = 0.25);
axes[0,0].tick_params(axis='x', rotation=45);


# Men Age in 2021
sns.barplot(ax=axes[0,1], data=age_man21, x='Age', y='Count', color='#ffab40');
axes[0,1].set_title('Men Age in 2021',color = 'lightpink', weight = 'bold');
axes[0,1].grid(axis='y', lw = 0.25);
axes[0,1].tick_params(axis='x', rotation=45);


# Women Age in 2020
sns.barplot(ax=axes[1,0], data=age_woman20, x='Age', y='Count', color='#ff4094');
axes[1,0].set_title('Women Age in 2020',color = 'lightpink', weight = 'bold');
axes[1,0].grid(axis='y', lw = 0.25);
axes[1,0].tick_params(axis='x', rotation=45);


# Women Age in 2021
sns.barplot(ax=axes[1,1], data=age_woman21, x='Age', y='Count', color='#ff4094');
axes[1,1].set_title('Women Age in 2021',color = 'lightpink', weight = 'bold');
axes[1,1].grid(axis='y', lw = 0.25);
axes[1,1].tick_params(axis='x', rotation=45);

plt.savefig('plot3.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
